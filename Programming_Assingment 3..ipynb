```python
#Q1.  Write a Python Program to Check if a Number is Positive, Negative or Zero?

Number = int(input())
if Number < 0:
    print("Negative")
    
elif Number >0:
    print("Positive")
else:
    print("Zero")
```

```python
#Q2.   Write a Python Program to Check if a Number is Odd or Even?
num = int(input())

if num %2 ==0:
    print("even")
else:
    print("Odd")
```

```python
#Q3.   Check the leap year or not.

year = int(input())
if (year%400==0) or (year %4 ==0 and year %100!=0):
    print("leap year")
else:
    print("Not leap year")
```

```python
#Q4.   Write a Python Program to Check Prime Number?

num = int(input())
if num <= 1:
    print("Not a Prime Number ")
else:
    for item in range(2,num):
        if num % item ==0:
            ("Not A Prime Number")
            break
    else:
        print("Prime Number")
```
